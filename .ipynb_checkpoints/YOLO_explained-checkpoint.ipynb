## __Forward Pass Through The Model__

```python
yolo_model_outputs = yolo_model(image_data)
print(f"YOLO Model Outputs{yolo_model_outputs.shape}")
yolo_outputs = yolo_head(yolo_model_outputs, anchors, len(class_names))
print(f"YOLO Model Outputs After Head : {len(yolo_outputs)}")
```

__As mentioned on the Blog, we have an output of dimensions 19 x 19 x 425, representing a grid of 19 x 19 cells and 5 predicted boxes for each cell__

## __Post Processing : Retrieving Boxes From The Output__

__We need to reshape the anchors to batch, height, width, num_anchors, box_params in order to match the outputs from the model. We also build a Conv Dims Tensor showing the 19 x 19 grid__

```python
num_anchors = len(anchors)
anchors_tensor = K.reshape(K.variable(anchors), [1, 1, 1, num_anchors, 2])
print(f"Anchors Tensor : {anchors_tensor}")
conv_dims = K.shape(yolo_model_outputs)[1:3]  # assuming channels last
print(f"Conv Dims : {conv_dims}")
```

__Then we construct the grid dimensions as pair-wise spacial points__

```python
conv_height_index = K.arange(0, stop=conv_dims[0])
conv_width_index = K.arange(0, stop=conv_dims[1])
print(f"Conv Height Index : {conv_height_index} , Conv Width Index : {conv_width_index}")
```

```python
conv_height_index = K.tile(conv_height_index, [conv_dims[1]])
conv_height_index
```

```python
conv_width_index = K.tile(K.expand_dims(conv_width_index, 0), [conv_dims[0], 1])
conv_width_index = K.flatten(K.transpose(conv_width_index))
conv_width_index
```

```python
conv_index = K.transpose(K.stack([conv_height_index, conv_width_index]))
conv_index[:10]
```

```python
conv_index = K.reshape(conv_index, [1, conv_dims[0], conv_dims[1], 1, 2])
conv_index = K.cast(conv_index, K.dtype(yolo_model_outputs))
conv_index[:,:,0,:]
```

__Feats is the reshaped version of the model outputs in order to identify each of the five boxes__

```python
feats = K.reshape(yolo_model_outputs, [-1, conv_dims[0], conv_dims[1], num_anchors, 80 + 5])
print(f"Feats Shape : {feats.shape}")
conv_dims = K.cast(K.reshape(conv_dims, [1, 1, 1, 1, 2]), K.dtype(feats))
print(f"Conv Dims : {conv_dims}")
```

## __Transforming Neural Network Box Outputs With The Help Of Anchor Boxes__

<img src="anchors.jpg" width="400" height="400">

__We generate some previous calculations__

```python
box_xy = K.sigmoid(feats[..., :2])
box_wh = K.exp(feats[..., 2:4])
box_confidence = K.sigmoid(feats[..., 4:5])
box_class_probs = K.softmax(feats[..., 5:])
```

## __XY Coordinates__

```python
print(f"Box XY Shape : {box_xy.shape}")
print(f"Conv Index Shape : {conv_index.shape}")
print(f"Conv Dims Shape : {conv_dims.shape}")
```

__We´ll work with cell (0,1). The first thing we do is to change the XY coordinates respect to the origin__

```python
box_xy_10 = box_xy[0][0][1]
print(f"box_xy_10 : {box_xy_10}")
conv_index_10 = conv_index[0][0][1]
print(f"conv_index_10 : {conv_index_10}")
box_xy_10 =  box_xy_10 + conv_index_10
print(f"box_xy_10 Coordinates After : {box_xy_10}")
```

__Then we calculate the coordinates as a proportion of the whole grid (height and width)__

```python
print(f"Conv Dims : {conv_dims}")
box_xy_10 = box_xy_10 / conv_dims
box_xy = (box_xy + conv_index) / conv_dims
print(f"box_xy_10 : {box_xy_10}")
```

## __HW Coordinates__

```python
print(f"Anchors Tensor : {anchors_tensor.shape}")
print(f"WH Tensor : {box_wh.shape}")
```

__We keep working with cell (0,1). We first rescale the Anchor Boxes and then calculate HW coordinates as a proportion of the whole grid__

```python
box_wh_10 = box_wh[0][0][1]
print(f"box_wh_10 : {box_wh_10}")
print(f"Anchors Tensor : {anchors_tensor}")
box_wh_10 = box_wh_10 * anchors_tensor
print(f"New Rescaled box_wh_10 : {box_wh_10}")
box_wh_10 = box_wh_10 / conv_dims
print(f"Final box_wh_10 : {box_wh_10}")
box_wh = box_wh * anchors_tensor / conv_dims
```

## __YOLO Building Boxes Part II__

### __YOLO Boxes to Corners__

__Again we work with the outputs from grid (0,1)__

```python
box_xy_10 = box_xy[0][0][1]
print(f"box_xy_10 : {box_xy_10}")
box_wh_10 = box_wh[0][0][1]
print(f"box_wh_10 : {box_wh_10}")
```

__We have XYWH, however we need to transform them into spacial points__

```python
box_mins_10 = box_xy_10 - (box_wh_10 / 2)
box_maxes_10 = box_xy_10 + (box_wh_10 / 2)
print(f"Box Mins : {box_mins_10}")
print(f"Box Maxes : {box_maxes_10}")
box_mins = box_xy - (box_wh / 2.)
box_maxes = box_xy + (box_wh / 2.)
```

__We concatenate everything__

```python
boxes_10 = tf.keras.backend.concatenate([
        box_mins_10[..., 1:2],  # y_min
        box_mins_10[..., 0:1],  # x_min
        box_maxes_10[..., 1:2],  # y_max
        box_maxes_10[..., 0:1]  # x_max
    ])
boxes = tf.keras.backend.concatenate([
        box_mins[..., 1:2],  # y_min
        box_mins[..., 0:1],  # x_min
        box_maxes[..., 1:2],  # y_max
        box_maxes[..., 0:1]  # x_max
    ])
boxes_10
```

### __YOLO Filtering Boxes__

__Moving on we need to filter the boxes which indeed have something on them based on a certain threshold__

```python
box_confidence_10 = box_confidence[0][0][1]
box_class_probs_10 = box_class_probs[0][0][1]
print(f"Box Confidence 10 : {box_confidence_10}")
print(f"Box Class Probs 10 80 probs for each of the 5 boxes : {box_class_probs_10.shape}")
```

```python
box_scores_10 = box_class_probs_10*box_confidence_10
box_scores = box_class_probs*box_confidence
```

```python
box_classes_10 = tf.math.argmax(box_scores_10,axis=-1)
box_class_scores_10 = tf.math.reduce_max(box_scores_10,axis=-1)
box_classes = tf.math.argmax(box_scores,axis=-1)
box_class_scores = tf.math.reduce_max(box_scores,axis=-1)
print(f"Clases con Mayor Score : {box_classes_10}")
print(f"Scores de las Clases con Mayores Scores : {box_class_scores_10}")
```

```python
threshold = .3
```

```python
filtering_mask_10 = (box_class_scores_10 >= threshold)
filtering_mask = (box_class_scores >= threshold)
print(f"Filtering Mask 10 : {filtering_mask_10}")
print(f"Box Class Scores Before Filtering Mask : {box_class_scores_10}")
scores_10 = tf.boolean_mask(box_class_scores_10,filtering_mask_10)
print(f"Box Class Scores After Filtering Mask : {scores_10}")
print(f"Boxes Before Filtering Mask : {boxes_10}")
boxes_10 = tf.boolean_mask(boxes_10,filtering_mask_10)
print(f"Boxes After Filtering Mask : {boxes_10}")
print(f"Box Classes Before Filtering Mask : {box_classes_10}")
classes_10 = tf.boolean_mask(box_classes_10,filtering_mask_10)
print(f"Box Classes After Filtering Mask : {classes_10}")
```

__We apply this to the whole set and we keep only the useful boxes__

```python
box_classes = tf.math.argmax(box_scores,axis=-1)
box_class_scores = tf.math.reduce_max(box_scores,axis=-1)
filtering_mask = (box_class_scores >= threshold)
scores = tf.boolean_mask(box_class_scores,filtering_mask)
boxes = tf.boolean_mask(boxes,filtering_mask)
classes = tf.boolean_mask(box_classes,filtering_mask)
```

```python
boxes
```

### __Rescaling Boxes__

__We need to remember that the outputs from YOLO are from a 19 x 19 grid, so we need to rescale them in order to draw the boxes. However since we were working with proportions, it´s really easy to do this process__

```python
image_shape = (720, 1280)
height = image_shape[0] * 1.0
width = image_shape[1] * 1.0
print(f"Image Height : {height} , Image Width : {width}")
```

```python
image_dims = K.stack([height, width, height, width])
image_dims = K.reshape(image_dims, [1, 4])
print(f"Image Dims : {image_dims}")
```

```python
boxes = boxes * image_dims
boxes
```

### __Non Max Supression__

__The core concept of Non Max Supression is to remove overlapping boxes and only leave the most representative one. In simple, you select the box with the biggest probability and then start comparing the overlapping ones. If this overlap, which is measured by the Intersection Over Union metric is greater that a certain threshold, you remove the box with the lowest probability__

```python
iou_threshold=.5
```

__Set Max Number of Boxes__

```python
max_boxes = 10
max_boxes_tensor = tf.Variable(max_boxes, dtype='int32') 
```

```python
nms_indices = tf.image.non_max_suppression(boxes,scores,max_boxes_tensor,iou_threshold)
```

```python
nms_indices
```

```python
scores = tf.gather(scores,nms_indices)
boxes = tf.gather(boxes,nms_indices)
classes = tf.gather(classes,nms_indices)
```
